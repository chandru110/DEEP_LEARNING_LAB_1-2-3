```python
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_derivative(x):
    return x*(1-x)

# Given Data
x = np.array([[1,0]])
y = np.array([[1]])

W1 = np.array([[0.2,0.3],
               [0.4,0.1]])

W2 = np.array([[0.5],
               [0.6]])

learning_rate = 0.1

# Forward Propagation
hidden_input = np.dot(x,W1)
hidden_output = sigmoid(hidden_input)

output_input = np.dot(hidden_output,W2)
prediction = sigmoid(output_input)

loss = 0.5*np.square(y-prediction)

print("Hidden Output:")
print(hidden_output)

print("\nPrediction:")
print(prediction)

print("\nLoss:")
print(loss)

# Backpropagation
output_delta = (prediction-y)*sigmoid_derivative(prediction)

hidden_delta = np.dot(output_delta,W2.T)*sigmoid_derivative(hidden_output)

dW2 = np.dot(hidden_output.T,output_delta)
dW1 = np.dot(x.T,hidden_delta)

W2_new = W2-learning_rate*dW2
W1_new = W1-learning_rate*dW1

print("\nGradient W2")
print(dW2)

print("\nGradient W1")
print(dW1)

print("\nUpdated W2")
print(W2_new)

print("\nUpdated W1")
print(W1_new)
```
